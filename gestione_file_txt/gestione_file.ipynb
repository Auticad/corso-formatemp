# Lavorare con i File

  <li>r - Read: apre il file per leggere, errore se non esiste</li>
  <li>a - Append: apre il file per appendere, crea il file se non esiste</li>
  <li>w - Write: apre il file per scrivere, crea il file se non esiste</li>
  <li>× - Create: crea il file, errore se esiste</li>
</ul>

```python
from pathlib import Path

path = Path("G:/AMBIENTE_VIRTUALE_PYTHON/FLASK/esempio_uno.txt")
```

```python
if path:
    print(f"Il file {nome_file} esiste.")
    
```

```python
file = open(path, "r")
contenuto = file.read()
print(contenuto)
file.close()
```

```python
with open(path, "r") as file:
    i = 1
    for line in file:
        print(line, end="")
        i += 1
```

```python
with open(path, "r") as file:
    line = file.readline()
    i = 1
    while line:
        print(f"{i}: {line}", end="")
        line = file.readline()
        i += 1
```

Scrittura di un file

```python
file_nuovo = open(path, "a")
file_nuovo.write("\nCiao, questo è un nuovo file.")
file_nuovo.close()
```

Esercizio 1 – Creare un DataFrame
Domanda

Crea un DataFrame con questi dati:

| prodotto | prezzo | quantità |
| -------- | ------ | -------- |
| Laptop   | 900    | 5        |
| Mouse    | 20     | 50       |
| Tastiera | 45     | 30       |
| Monitor  | 300    | 10       |


Poi:
Mostra le informazioni del DataFrame

```python
prodotti = [
    {"prodotto": "Laptop", "prezzo": 900, "quantità": 5},
    {"prodotto": "Mouse", "prezzo": 20, "quantità": 50},
    {"prodotto": "Tastiera", "prezzo": 45, "quantità": 30},
    {"prodotto": "Monitor", "prezzo": 300, "quantità": 10}
]
```

```python
import pandas as pd

df_prodotti = pd.DataFrame(prodotti)

df_prodotti
```

Esercizio 2 – Calcolare il totale delle vendite
Domanda

Crea una nuova colonna chiamata totale:

totale = prezzo * quantità

Poi stampa il DataFrame

```python
df_prodotti["totale"] = df_prodotti["prezzo"] * df_prodotti["quantità"]
df_prodotti
```

Esercizio 3 – Statistiche
Domanda

Calcola:
<ol type="1">
  <li>media dei prezzi</li>
  <li>prezzo massimo</li>
  <li>prezzo minimo</li>
</ol>

```python
media =df_prodotti["totale"].mean()
print(f"Il totale medio è: {media}")

massimo =df_prodotti["totale"].max()
print(f"Il totale massimo è: {massimo}")

minimo = df_prodotti["totale"].min()
print(f"Il totale minimo è: {minimo}")
```

Esercizio 4 – Grafico
Domanda

Crea un grafico a barre con:
<ul style="list-style-type:disc;">
  <li>prodotti sull'asse X</li>
  <li>quantità sull'asse Y</li>
</ul>

```python
from matplotlib import pyplot as plt

df_prodotti.plot(kind="bar", x="prodotto", y="totale", title="Totale per Prodotto")
plt.xlabel("Prodotto")
plt.ylabel("Totale")
plt.show()
```

Esercizio 5 – Analisi dati
Domanda

Rispondi alle domande:
<ul style="list-style-type:disc;">
  <li>Qual è il prodotto più costoso?</li>
  <li>Qual è il prodotto venduto di più?</li>
  <li>Qual è il totale delle vendite?</li>
</ul>

```python
più_costoso = df_prodotti["prezzo"].max()
print(f"Il prodotto più costoso ha un prezzo di: {più_costoso}")

più_venduto = df_prodotti["quantità"].max()
print(f"Il prodotto più venduto ha una quantità di: {più_venduto}")

tot_venduto = df_prodotti["totale"].sum()
print(f"Il totale venduto è: {tot_venduto}")
```

```python
# Opzione A — usi la colonna 'numeri' dopo averla parsata
import ast

def normalizza(serie: pd.Series) -> pd.Series:
    minv = serie.min()
    maxv = serie.max()
    return (serie - minv) / (maxv - minv)   # ← breakpoint qui

df['numeri_parsed'] = df['numeri'].apply(ast.literal_eval)
#df['media'] = df['numeri_parsed'].apply(lambda x: sum(x) / len(x))
df['norm'] = normalizza(df['numeri_parsed'].apply(lambda x: sum(x) / len(x)))

df.head()
```

```python
from matplotlib import pyplot as plt


data = [5,15,12,17,15,26,25,39,30]
plt.hist(data, bins=5, edgecolor='black', color='skyblue')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [15,18,25,30,34]
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [45,25,15,15]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Programming Languages')
plt.show()
```

```python
x = [5,7,9,2,12,17]
y = [99,87,56,97,46,78]
plt.scatter(x,y, color='green')
plt.title('Scatter Plot')
plt.show()
```

```python
names = ["A","B","C","D"]
values = [10,20,15,25]
plt.bar(names, values, color='purple')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

<h1>Parametri comuni in Pandas / Matplotlib</h1>


| Parametro           | Dove si usa                                                      | Significato                                           | Esempio                                                                 |
| ------------------- | ---------------------------------------------------------------- | ----------------------------------------------------- | ----------------------------------------------------------------------- |
| `kind`              | `df.plot()`                                                      | Tipo di grafico                                       | `"bar"`, `"line"`, `"hist"`, `"pie"`                                    |
| `x`                 | `df.plot()`                                                      | Colonna da usare per asse X                           | `x="nome"`                                                              |
| `y`                 | `df.plot()`                                                      | Colonna da usare per asse Y                           | `y="voto"`                                                              |
| `color`             | `df.plot()` o matplotlib                                         | Colore barre o linee                                  | `color="orange"`                                                        |
| `figsize`           | `df.plot()` o matplotlib                                         | Dimensione grafico in pollici                         | `figsize=(10,6)`                                                        |
| `autopct`           | `plot(kind="pie")`                                               | Mostra percentuale sulle fette della torta            | `autopct='%1.1f%%'` → 1 cifra decimale, con `%`                         |
| `startangle`        | `plot(kind="pie")`                                               | Rotazione iniziale della torta in gradi               | `startangle=90` → inizia dall’alto                                      |
| `inplace`           | Funzioni DataFrame (`fillna()`, `drop_duplicates()`, `rename()`) | Modifica il DataFrame direttamente senza creare copia | `df.drop_duplicates(inplace=True)`                                      |
| `ascending`         | `sort_values()`                                                  | Ordina crescente o decrescente                        | `df.sort_values("voto", ascending=False)` → decrescente                 |
| `bins`              | `plot(kind="hist")`                                              | Numero di intervalli per istogramma                   | `df["voto"].plot(kind="hist", bins=5)`                                  |
| `edgecolor`         | `plot(kind="hist")`                                              | Colore contorno barre istogramma                      | `edgecolor="black"`                                                     |
| `ylabel`            | matplotlib                                                       | Etichetta asse Y                                      | `plt.ylabel("Voto")`                                                    |
| `xlabel`            | matplotlib                                                       | Etichetta asse X                                      | `plt.xlabel("Studente")`                                                |
| `title`             | matplotlib                                                       | Titolo grafico                                        | `plt.title("Voti studenti")`                                            |
| `groupby()`         | DataFrame                                                        | Raggruppa dati per colonna                            | `df.groupby("corso")["voto"].mean()`                                    |
| `apply()`           | DataFrame / Serie                                                | Applica funzione a ogni elemento                      | `df["esito"] = df["voto"].apply(lambda x: "Pass" if x>=27 else "Fail")` |
| `value_counts()`    | Serie                                                            | Conta quante volte compare ciascun valore             | `df["corso"].value_counts()`                                            |
| `reset_index()`     | DataFrame                                                        | Ricostruisce indice dopo `groupby` o ordinamento      | `df.groupby("corso")["voto"].max().reset_index()`                       |
| `fillna()`          | DataFrame / Serie                                                | Riempie valori mancanti                               | `df["voto"].fillna(df["voto"].mean(), inplace=True)`                    |
| `dropna()`          | DataFrame / Serie                                                | Elimina righe con valori mancanti                     | `df.dropna(inplace=True)`                                               |
| `drop_duplicates()` | DataFrame                                                        | Elimina righe duplicate                               | `df.drop_duplicates(inplace=True)`                                      |
| `astype()`          | DataFrame / Serie                                                | Cambia tipo di dati colonna                           | `df["voto"] = df["voto"].astype(int)`                                   |
| `head()`            | DataFrame                                                        | Mostra prime N righe                                  | `df.head(5)`                                                            |
| `tail()`            | DataFrame                                                        | Mostra ultime N righe                                 | `df.tail(5)`                                                            |
| `mean()`            | Serie                                                            | Calcola media                                         | `df["voto"].mean()`                                                     |
| `max()`             | Serie / DataFrame                                                | Calcola massimo                                       | `df["voto"].max()`                                                      |
| `min()`             | Serie / DataFrame                                                | Calcola minimo                                        | `df["voto"].min()`                                                      |

<h3>Creare la lista dei voti: votoId-studenteId-materiaId-voto</h3>

```python
voti = [
    {"votoId":1, "studenteId":1, "materiaId":1, "voto":8},
    {"votoId":2, "studenteId":1, "materiaId":2, "voto":9},
    {"votoId":3, "studenteId":1, "materiaId":3, "voto":7},
    {"votoId":4, "studenteId":2, "materiaId":1, "voto":6},
    {"votoId":5, "studenteId":2, "materiaId":2, "voto":8},
    {"votoId":6, "studenteId":2, "materiaId":3, "voto":9},
    {"votoId":7, "studenteId":3, "materiaId":1, "voto":7}
]

df_voti = pd.DataFrame(voti)

print(df_voti)

path = Path("G:/AMBIENTE_VIRTUALE_PYTHON/FLASK/voti.csv")
df_voti.to_csv(path, index=False)
```

```python
media = df_voti['voto'].mean()
print("La media è: ", media)
```

```python
media_voti = df_voti.groupby('materiaId')['voto'].mean().reset_index()
print(media_voti)
```

```python
media_voti.plot(x='materiaId', y='voto', kind='line', marker='o', markersize=10, color='blue', linewidth=2  )
plt.title('Media dei voti per materia')
plt.xlabel('materia')
plt.ylabel('voto')
plt.show()
```

```python
df_voti["voto"].plot(kind='line', marker='o', markersize=10, color='blue', linewidth=2)
plt.title('Media dei voti per materia')
plt.xlabel('voto')
plt.ylabel('numero studenti')
plt.show()
```

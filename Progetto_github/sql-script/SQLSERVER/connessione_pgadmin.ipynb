```python
import psycopg2

try:
    # Connessione (inserisci le tue credenziali)
    conn = psycopg2.connect(
        host="localhost",
        database="postgres",
        user="postgres",
        password="12345678",
        port="5432"
    )
    
    # Creazione del cursore
    cur = conn.cursor()
    
    # Query standard per ottenere i nomi delle tabelle
    query = """
    SELECT table_name 
    FROM information_schema.tables 
    WHERE table_schema = 'public' 
    AND table_type = 'BASE TABLE';
    """
    
    cur.execute(query)
    tabelle = cur.fetchall()
    
    print(f"Tabelle trovate nel database 'postgre':")
    if not tabelle:
        print("- Nessuna tabella presente nello schema public.")
    for tab in tabelle:
        print(f" - {tab[0]}")
        
    # Chiusura risorse
    cur.close()
    conn.close()

except Exception as e:
    print(f"Errore durante la lettura del database: {e}")
```

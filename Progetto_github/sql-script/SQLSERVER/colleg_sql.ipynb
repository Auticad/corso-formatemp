```python
import os
import pyodbc
from dotenv import load_dotenv

# 1. Carica le variabili
load_dotenv()

# 2. Recupera i valori
server = os.getenv("DB_SERVER", "localhost")
database = os.getenv("DB_NAME")
driver = os.getenv("DB_DRIVER", "ODBC Driver 17 for SQL Server")
trusted = os.getenv("TRUSTED_CONNECTION", "no").lower()
trust_cert = os.getenv("TRUST_SERVER_CERTIFICATE", "no").lower()

# 3. Costruisci la stringa di connessione in modo dinamico
# Iniziamo con i parametri base
conn_str = (
    f"DRIVER={{{driver}}};"
    f"SERVER={server};"
    f"DATABASE={database};"
)

# Aggiungiamo TrustServerCertificate se necessario
if trust_cert == "yes":
    conn_str += "TrustServerCertificate=yes;"

# Gestiamo l'autenticazione (Windows vs SQL Server)
if trusted == "yes":
    # Autenticazione Windows: NON inviare UID e PWD
    conn_str += "Trusted_Connection=yes;"
else:
    # Autenticazione SQL: invia utente e password
    username = os.getenv("DB_USER")
    password = os.getenv("DB_PASSWORD")
    conn_str += f"UID={username};PWD={password};"

# 4. Tentativo di connessione
try:
    print(f"🔍 Tentativo di connessione con:\n{conn_str}\n")
    conn = pyodbc.connect(conn_str)
    print("✅ Connessione avvenuta con successo!")
except Exception as e:
    print(f"❌ Errore di connessione: {e}")
```
